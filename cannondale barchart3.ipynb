```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import functools
import operator
import math

f = open('cannondaleResultsCombined.json')
 
data = json.load(f)

hashtagList = []
likeList = []
filteredHashList = []


#for post in data:
#    print(post["likes"])
#    print(post["hashtags"])

for post in data:
     likes = post["likes"]
     hashes = post["hashtags"]
     # Change like threshold depending on how many posts wanted in chart
     if likes > 30000:
        forDataFrame = {
            'Likes': likes,
            'Hashtags': hashes
        }
        likeList.append(list(forDataFrame.values())[0])
        hashtagList.append(list(forDataFrame.values())[1])
        

y = likeList
x = hashtagList

print(len(x))
#print(type(y))
print(len(y))
#print(x)
#print(y)


df = pd.DataFrame({'hashtags':x, 'likes':y})
ax = df.plot.bar(x='hashtags', y='likes', rot=90, figsize=(16,9))

```
